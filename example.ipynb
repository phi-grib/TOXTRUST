```python
from combine.libraries import *
from combine.evidence_combinator import SingleEvidence, EvidenceCombinator
```

### Single class objects

```python
endpoint_invitro = SingleEvidence(identifier = 'experiment', source = 'in vitro', result = 1, reliability = 0.85, relevance='plausible')
```

```python
endpoint_insilico = SingleEvidence(identifier = 'model', source = 'QSAR', result = [0.3033339, 0.6966661 ], reliability = [0.65,0.74], relevance='certain')
```

```python
endpoint_insilico.VisualiseSingleEvidence()
```

### Combination 

```python
example_compound = EvidenceCombinator('Example Endpoint', 'Example Compound')
```

```python
example_compound.AddEvidence(endpoint_invitro)
example_compound.AddEvidence(endpoint_insilico)
```

```python
example_compound.Combination(rule_selection='Dempster', WoE=False)
```

```python
example_compound.Visualise() 
```

```python
example_compound.ReturnResults(selection='Dempster')
```
